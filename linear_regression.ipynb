```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Provide data
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])
```

```python
x
```

```python
y
```

```python
# Create a model and fit it
model = LinearRegression()
```

```python
model.fit(x, y)
```

```python
model = LinearRegression().fit(x, y)
```

```python
r_sq = model.score(x, y)
print(f"coefficient of determination: {r_sq}")
```

```python
print(f"intercept: {model.intercept_}")
```

```python
print(f"slope: {model.coef_}")
```

### Another Example

```python
# importing models
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
```

```python
url = 'https://gist.githubusercontent.com/jiisanda/b3c617053f2338d5f29fa75715cf9141/raw/3aeed18f15466542e6627224a0508da4ed61794f/dataset.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
# Data Validation
# Shape
print("Number of columns:", df.shape[1])
print("Number of rows: ", df.shape[0])
# info
df.info()
# description
df.describe()
```

```python
plt.scatter(x = df['YearsExperience'], y = df['Salary'], color = 'blue', edgecolor='black', linewidth=1)

plt.title('Salary w.r.t years of experience')
plt.xlabel('Years of Expericne')
plt.ylabel('Salary')


plt.tight_layout()
plt.show
```

### 2) Data Preprocessing

```python
# Define Predictor and target Attribute
X = df.iloc[:, :-1]
Y = df.iloc[:, -1]
```

```python
# Splitting the data to training and testing
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
# Training Data Set
print("X_test", X_train.shape)
print("X_test", y_train.shape)
```

```python
# Testing Data Set
print("X_test", X_test.shape)
print("X_test", y_test.shape)
```

3) Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

4) Evaluate Result

```python
# Visualizing the result
plt.scatter(X_test, y_test, color = 'blue', edgecolor='black', linewidth=1)
plt.plot(X_test, y_pred, color = 'red', linewidth = 1)

plt.title('Salary w.r.t years of experience')
plt.xlabel('Years of Expericne')
plt.ylabel('Salary')


plt.tight_layout()
plt.show
```
